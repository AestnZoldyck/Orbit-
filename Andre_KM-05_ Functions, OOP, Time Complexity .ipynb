## **Inheritence**

```python
class Bata(object):
  def __init__(self, nama, usia, warna, jenis):
    self.nama = nama
    self.usia = usia
    self.warna = warna
    self.jenis = jenis

  def bakar(self, durasi):
    for x in range(durasi):
      print("wusshh.... gubrak.... wuussshh.... ")

  def info(self):
    print(f"Nama: {self.nama}, Usia: {self.usia},Warna: {self.warna}, Jenis: {self.jenis}")


Bata1 = Bata("Bata",2, "Merah","Tanah")

Bata1.info()
Bata1.bakar(8)

class Batu(Bata):
  def __init__(self,  nama, usia, warna, jenis_Batu):
    super().__init__(nama, usia, warna,jenis_Batu)
    self.warna = warna
    self.jenis_Batu = jenis_Batu
  def bakar2(self, durasi):
    for x in range(durasi):
      print("wusshh.... wuussshh.... ")

  def info_Batu(self):
    print(f"jenis : {self.jenis_Batu}")

Batu = Batu("Bata", 4 , "Merah", "Tanah")

Batu.info()
print('')
Batu.bakar2(8)
Batu.info_Batu()
```

```python
# 11

# algoritma binary search
# NIlai dari variabel nilai_a dan nilai_b yang dicari didalam list, bisa dirubah sesuai kebutuhan untuk testing hasil lainnya

nilai_a = 17
nilai_b = 3

def bin_search(ourlist, key):
    left = 0 # I assign left position to zero
    right = len(ourlist) # I assign right position by defining the length of ourlist minus one 
    matched = True
    while(left<=right and not matched): # the loop will continue untill the left element is less 
                                        # or equal to the right element and the matched is True
        nilai= (left+right)//2 # I find the position of the middle element
        if ourlist[nilai_a] == key: # if the middle element correponds to the key element
             matched = False
        else: #otherwise 
            if key < ourlist[nilai_b]: # if key element is less than the middle element
                right = nilai_b #I assign the position of the right element as mid - 1
            else: #otherwise
                left = nilai_a  #left position will become the middle position plus 1
    return matched


print(bin_search([1, 3, 9, 15, 17], nilai_a))
print(bin_search([1, 3, 9, 15], nilai_b))
```
